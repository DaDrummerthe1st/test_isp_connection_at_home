```python
# function to easy up the process
# Create a 3-hour chunk of datetimes
# Labels must be exactly one fewer than edges (boundry between bins)
# 1-3,3-6,6-9 ==> 
# starting values (one "side" of edges): 1,3,6
# end values (the other side of edges): 3,6,9
# so edges are then:  1,3,6,9
def create_time_bins(df: object, success: bool, bin_size: int):
    bins = range(0, 25, bin_size)
    labels = [f"{i:02d}:00-{i+bin_size:02d}:00" for i in bins[:-1]]
    df_filtered = df.loc[df['success'] == success].copy()
    df_filtered.loc[:, 'time_bin']  = pd.cut(df_filtered['timestamp'].dt.hour,
                                             bins=bins,
                                             labels=labels,
                                             right=False)
    #df_filtered = df_filtered.groupby('time_bin', observed=False).size()
    return df_filtered

# Failures
failures_per_bin_hourly = create_time_bins(df, 0, 1)
failures_per_bin_three_hours = create_time_bins(df, 0, 3)

# successes
successes_per_bin_hourly = create_time_bins(df, 1, 1)
successes_per_bin_three_hours = create_time_bins(df, 1, 3)

display(successes_per_bin_hourly)
```
